```python
def count_punct(text):
    """
    Подсчитывает процент символов пунктуации в тексте.

    Параметры:
    text (str): Текст, в котором необходимо подсчитать символы пунктуации.

    Возвращает:
    float: Процент символов пунктуации в тексте
    """
    count = sum([1 for char in text if char in string.punctuation])
    return round(count/(len(text) - text.count(" ")), 3) * 100

def train_model_tf_idf(model, param_grid):
    """
    Обучает модель с использованием заданных гиперпараметров и оценивает ее производительность.

    Параметры:
    model (estimator): Модель для обучения.
    param_grid (dict): Сетка гиперпараметров для поиска наилучших гиперпараметров.

    Возвращает:
    tuple: Кортеж, содержащий оценку ROC AUC на кросс-валидации, время предсказания и лучший обученный пайплайн модели.
    """
    pipeline = Pipeline([
        ('preprocessor', ColumnTransformer([
            ('tfidf', TfidfVectorizer(), 'lemm_text')
        ], remainder='passthrough')),
        ('model', model)
    ])

    grid_search = RandomizedSearchCV(pipeline, param_grid, cv=3, scoring='roc_auc')
    grid_search.fit(features_train, target_train)

    best_pipeline = grid_search.best_estimator_

    roc_auc_cv = grid_search.best_score_

    start_time = time.time()
    predictions = best_pipeline.predict(features_train)
    end_time = time.time()
    pred_time = end_time - start_time

    print("ROC AUC на кросс-валидации:", roc_auc_cv)
    print("Время предсказания:", pred_time, "секунд")

    return roc_auc_cv, pred_time, best_pipeline

def train_model_bert(model, param_grid):
    """
    Обучает модель с использованием заданных гиперпараметров и оценивает ее производительность.

    Параметры:
    model (estimator): Модель для обучения.
    param_grid (dict): Сетка гиперпараметров для поиска наилучших гиперпараметров.

    Возвращает:
    tuple: Кортеж, содержащий оценку ROC AUC на кросс-валидации, время предсказания и лучший обученный пайплайн модели.
    """
    pipeline = Pipeline([
          ('model', model)
    ])


    grid_search = RandomizedSearchCV(pipeline, param_grid, cv=3, scoring='roc_auc')
    grid_search.fit(features_train_bert, target_train_bert)

    best_pipeline = grid_search.best_estimator_

    roc_auc_cv = grid_search.best_score_

    start_time = time.time()
    predictions = best_pipeline.predict(features_train_bert)
    end_time = time.time()
    pred_time = end_time - start_time

    print("ROC AUC на кросс-валидации:", roc_auc_cv)
    print("Время предсказания:", pred_time, "секунд")

    return roc_auc_cv, pred_time, best_pipeline

def clear_text(text):
    """
    Очищает текст от специальных символов и приводит к нижнему регистру.

    Аргументы:
    text (str): Исходный текст.

    Возвращает:
    str: Очищенный текст.
    """
    text = text.lower()
    clear_text = re.sub(r'[^a-zA-Z]', ' ', text)
    clear_text = " ".join(clear_text.split())

    return clear_text

def get_wordnet_pos(word):
    """
    Определяет часть речи слова в формате WordNet.

    Параметры:
    word (str): Слово, для которого нужно определить часть речи.

    Возвращает:
    str: Строка, представляющая часть речи в формате WordNet.
    """
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                    "N": wordnet.NOUN,
                    "V": wordnet.VERB,
                    "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

def lemmatize_text(text):
    """
    Лемматизирует текст, удаляет стоп-слова и возвращает обработанный текст.
    (Пришлось костылить функцую из-за некоректной работы Pandarallel с windows)

    Аргументы:
    text (str): Исходный текст.

    Возвращает:
    str: Обработанный текст после лемматизации и удаления стоп-слов.
    """
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens if token.lower() not in stop_words]

    return ' '.join(lemmatized_tokens)
```

## Transformers

Попробуем воспользоваться трансформерами, а именно предварительно обученной моделью "distilbert-base-uncased-finetuned-sms-spam-detection".







```python
model_name = "mariagrandury/distilbert-base-uncased-finetuned-sms-spam-detection"
model = AutoModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)

batch_size = 100

tokenized = train['text'].apply(
    lambda x: tokenizer.encode(x, add_special_tokens=True, max_length=512, truncation=True))

max_len = max(len(x) for x in tokenized)

padded = np.array([i + [0]*(max_len - len(i)) for i in tokenized.values])

attention_mask = np.where(padded != 0, 1, 0)

num_batches = math.ceil(padded.shape[0] / batch_size)

embeddings = []

for i in tqdm(range(num_batches)):
    start_idx = batch_size * i
    end_idx = min(batch_size * (i + 1), padded.shape[0])
    batch = torch.LongTensor(padded[start_idx:end_idx]).to(device)
    attention_mask_batch = torch.LongTensor(attention_mask[start_idx:end_idx]).to(device)

    with torch.no_grad():
        batch_embeddings = model(batch, attention_mask=attention_mask_batch)

    embeddings.append(batch_embeddings[0][:,0,:].detach().cpu().numpy())
```

```python
features = np.concatenate(embeddings)
target = train['text_type']

features_train_bert, features_valid_bert, target_train_bert, target_valid_bert = train_test_split(
    features, target, test_size=0.25, random_state=RANDOM_STATE)
```

```python
model = LogisticRegression(random_state=RANDOM_STATE)
param_grid = {
     'model__C': [0.01, 0.1, 1, 10]
}

roc_auc_lr_bert, pred_time_lr_bert, best_model_lr_bert = train_model_bert(model, param_grid)
```

```python
model = RandomForestClassifier(random_state=RANDOM_STATE)
param_grid = {
     'model__n_estimators': [100, 200],
     'model__max_depth': [None, 10, 20]
}

roc_auc_rf_bert, pred_time_rf_bert, best_model_rf_bert = train_model_bert(model, param_grid)
```

```python
model = CatBoostClassifier(random_state=RANDOM_STATE, task_type="GPU", devices='0:1', verbose=0)
param_grid = {
     'model__iterations': [100, 200],
     'model__depth': [3, 6, 9]
}

roc_auc_cb_bert, pred_time_cb_bert, best_model_cb_bert = train_model_bert(model, param_grid)
```
